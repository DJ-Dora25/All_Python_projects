```python
# Conveting into date format.
df['Date']= pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
# reshaping the dataframe?
reshaped_df = df.pivot(index = "Date",columns = 'Tags',values = 'Posts')
```

```python
reshaped_df.shape
```

```python
reshaped_df.columns
```

```python
reshaped_df.count()
```

```python
reshaped_df.fillna(0, inplace = True)
```

```python
reshaped_df.head()
```

```python
reshaped_df.isna().values.any()
```

```python
# Compairing the rank between java and python over time?
plt.figure(figsize = (10,5))
plt.xlabel("Date")
plt.ylabel("Posts")
plt.plot(reshaped_df.index,reshaped_df.java)
plt.plot(reshaped_df.index,reshaped_df.python)
plt.ylim(0,40000)
```

```python
# To plot all the programming languages on the same chart?
plt.figure(figsize = (10,10))
plt.ylim(0,40000)

for columns in reshaped_df.columns:
    plt.plot(reshaped_df.index, reshaped_df[columns],label = reshaped_df[columns].name,linewidth = 3)
    plt.legend()
```

```python
roll_df = reshaped_df.rolling(window = 12).mean()
plt.figure(figsize = (10,10))
plt.xlabel('Date')
plt.ylabel('Number of Posts')
plt.ylim(0,40000)

#plot the roll_df:-
for columns in roll_df.columns:
    plt.plot(roll_df.index,roll_df[columns],label = roll_df[columns].name,linewidth = 3)
plt.legend()
```
